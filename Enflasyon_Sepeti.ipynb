# IMPORTLAR

```python
import pandas as pd
import numpy as np

import sys
import warnings

if not sys.warnoptions:
    warnings.simplefilter("ignore")
```

```python
data=pd.read_csv("data/food_data.csv")
data.head()
```

```python
#price kolonu object olarak göründüğü için tipini floata çeviyoruz.

data['fiyat'] = pd.to_numeric(data['fiyat'], errors = 'coerce') 
```

# Metin Verisini Düzenleme

```python
dic = {
    "ı":"i",
    "İ":"i",
    "ç":"c",
    "ö":"o",
    "ü":"u",
    "ğ":"g",
    "ş":"s"
}
```

```python
def replace_all(text, dic):
    for i, j in dic.items():
        text = text.replace(i, j)
    return text
```

```python
data["urun"] = data["urun"].apply(lambda x:replace_all(x.lower(),dic))
```

# MODEL

```python
data = data[['urun', 'madde']]
```

```python
from sklearn.utils import shuffle
data = shuffle(data)
data.reset_index(drop=True, inplace=True)
data.tail()
```

```python
df = data.iloc[:300000] # Model eğitimi için 300 bin yeterli
```

## BAG OF WORDS

```python
y=df["madde"].values
x=df["urun"].values
```

```python
#Bag of words yöntemini kullanmak için sklearn CountVectorizer fonksiyonunu kullanacağız.

from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(ngram_range=(1, 2)) #ngram kelimelerimizi 2 kelimelik gruplar halinde incelememize yarıyor.

bag = vectorizer.fit_transform(x)
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(bag,y)

y.shape,x.shape
```

## RandomForest

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_jobs=-1)
clf.fit(X_train,y_train)
```

## KNN

```python
# from sklearn.neighbors import KNeighborsClassifier

# neigh = KNeighborsClassifier(n_neighbors=3)
# neigh.fit(X_train,y_train)
```

## Prediction

```python
preds = clf.predict(X_test)
print(preds)
```

```python
# INVERSE PRED
```

```python
vectorizer.inverse_transform(X_test)
```

## Accuracy

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(preds, y_test)
print(accuracy)
```
