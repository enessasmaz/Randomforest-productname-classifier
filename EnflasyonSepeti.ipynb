# BAG OF WORDS

```python
y=df["madde"].values
x=df["prod"].values
```

```python
x
```

```python
y
```

```python
#Bag of words yöntemini kullanmak için sklearn CountVectorizer fonksiyonunu kullanacağız.

from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer(ngram_range=(1, 2)) #ngram kelimelerimizi 2 kelimelik gruplar halinde incelememize yarıyor.
```

```python
bag = vectorizer.fit_transform(x)
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
# from sklearn import preprocessing
# le = preprocessing.LabelEncoder()
# target=le.fit(y)
```

```python
# le.classes_
```

```python
# ye=le.transform(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(bag,y)
```

```python
y.shape,x.shape
```

# Model RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier(n_jobs=-1)
```

```python
clf.fit(X_train,y_train)
```

# Prediction

```python
preds = clf.predict(X_test)
```

```python
preds
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cf_matrix=confusion_matrix(preds,y_test)
```

```python
cf_matrix
```

# ACCURACY

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(preds, y_test)
```
